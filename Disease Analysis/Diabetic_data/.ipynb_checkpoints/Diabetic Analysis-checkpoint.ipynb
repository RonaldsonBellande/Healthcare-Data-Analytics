```python
#convert all features to numerical value
labelencoder_Y_axis = LabelEncoder()
Y_axis = labelencoder_Y_axis.fit_transform(Y_axis)
```

```python
#Replace <30 with 1. Rest of the values are 0
where_0 = np.where(Y_axis == 0)
where_1 = np.where(Y_axis == 1)

Y_axis[where_0] = 1
Y_axis[where_1] = 0
```

```python
#dataframe = pd.DataFrame(X_axis)
dataframe = pd.DataFrame(X_axis)
X_axis
where_0 = np.where(X_axis == '[0-10)')
where_1 = np.where(X_axis == '[10-20)')
where_2 = np.where(X_axis == '[20-30)')
where_3 = np.where(X_axis == '[30-40)')
where_4 = np.where(X_axis == '[40-50)')
where_5 = np.where(X_axis == '[50-60)')
where_6 = np.where(X_axis == '[60-70)')
where_7 = np.where(X_axis == '[70-80)')
where_8 = np.where(X_axis == '[80-90)')
where_9 = np.where(X_axis == '[90-100)')


X_axis[where_0] = 1
X_axis[where_1] = 2
X_axis[where_2] = 3
X_axis[where_3] = 4
X_axis[where_4] = 5
X_axis[where_5] = 6
X_axis[where_6] = 7
X_axis[where_7] = 8
X_axis[where_8] = 9
X_axis[where_9] = 10
```

```python
print(X_axis)
print(Y_axis)
# Spliting the dataset into the Training set and the Test set
X_train, X_test, Y_train, Y_test = train_test_split(X_axis, Y_axis, test_size = 0.20, random_state = 0)
```

```python
#convert all features to numerical value
labelencoder_X_axis = LabelEncoder()
X_axis = labelencoder_X_axis.fit_transform(X_axis)
```

```python
#Feature Scaling
standard_scaler = StandardScaler()
#Split it so you can train and test from diffrent sets
X_train = X_train.reshape(1, -1)
X_test = X_test.reshape(1, -1)
X_train = standard_scaler.fit_transform(X_train)
X_test = standard_scaler.transform(X_test)
```

```python
#Fitting the logistic Regression Algorithm to the Training set
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, Y_train)
```

```python
#Predicte the test set results
Y_pred = classifier.predict(X_test)
```

```python
#Create the confusion matrix
cm = confusion_matrix(Y_test, Y_pred)
print("Accuracy of Logistic Regression:", accuracy_score(Y_test, Y_pred))
```

```python
print("Precision of Logistic Regression:", precision_score(Y_test, Y_pred, average='weighted'))
```

```python
#Fitting the Support Vector Machines Algorithm to the Training set
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, Y_train)
```

```python
#Predict the test result
Y_pred = classifier.predict(X_test)
```

```python
#Create a confusion matrix
cm = confusion_matrix(Y_test, Y_pred)
print("Accuracy of SVC:", accuracy_score(Y_test, Y_pred))
```

```python
print("Precision of SVC:", precision_score(Y_test, Y_pred, average='weighted'))
```
